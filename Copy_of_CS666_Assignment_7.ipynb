**Task 4**: Let's apply a 3x3 filter mask! [30 Points]

```python
# Here we will modify the pixel data!
#
# The following code allows you to set the pixels of the current slice.
#
#   pixels.set(new_pixels);
#
# Write code to APPLY A FILTER MASK to the pixels.
#
# Hint: You can use the following code to get the dimensions of the current slice.
#   w = element.image.width;
#   h = element.image.height;
#
# You can decide which filter mask to apply (e.g., Sobel, Gaussian blur, edge detection..).
#
# After setting new pixel values, you can re-draw/re-render the current slice with this command:
#
#   cornerstone.renderGrayscaleImage(element, true);
#
```

```python
#
#
# TODO: YOUR CODE GOES HERE
#  USING EDGE DETECTION : 
let w = element.image.width;
let h = element.image.height;


let new_pixels = new Uint16Array(pixels.length);

for (let y = 1; y < h - 1; y++) {
  for (let x = 1; x < w - 1; x++) {
    let index = y * w + x;

    // Sobel operation
    let gaussx = (
      pixels[index - w - 1] +
      2 * pixels[index - 1] +
      pixels[index + w - 1] -
      pixels[index - w + 1] -
      2 * pixels[index + 1] -
      pixels[index + w + 1]
    );

    let gaussy = (
      pixels[index - w - 1] +
      2 * pixels[index - w] +
      pixels[index - w + 1] -
      pixels[index + w - 1] -
      2 * pixels[index + w] -
      pixels[index + w + 1]
    );

    // Calculating the value of the gradient
    let value = Math.sqrt(gaussx * gaussx + gaussy * gaussy);

    // Set the new pixel value
    new_pixels[index] = Math.round(value);
  }
}

// Setting the modified pixel values
pixels.set(new_pixels);

// Re-rendering the current slice
cornerstone.renderGrayscaleImage(element, true);
#
#
```
